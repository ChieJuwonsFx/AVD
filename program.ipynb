```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diabetes_binary_health_indicators_BRFSS2015.csv')
df
```

```python
df.info()

df.describe()

missing_values = df.isnull().sum()
missing_values
```

```python
df.describe()
```

```python
df.head()
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(),annot=True , cmap ='coolwarm' )
plt.title("correlation of feature")
```

```python
hp = df['Veggies']==1
veggies = df[hp]
veggies
```

```python
labels = ['DIABETES', 'NOT DIABETES']
sizes = [len(veggies[df.Diabetes_binary == 1]), len(veggies[df.Diabetes_binary == 0])]

colors = ['#FF7F7F', '#7FB3FF']
explode = (0.05, 0)  

fig, ax = plt.subplots(figsize=(10, 8), dpi=80)
patches, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.2f%%',
    pctdistance=0.75,
    startangle=140,
    textprops={'fontsize': 14, 'fontweight': 'bold', 'fontname': 'Poppins'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

for text in texts:
    text.set_color('#333333')
for autotext in autotexts:
    autotext.set_color('#444444')
    
plt.suptitle("Diabetes Throught Veggies",
           fontname='Poppins', fontsize=24, fontweight = 'bold')
plt.title("Perbandingan antara Konsumsi Sayur terhadap Diabetes",fontname='Poppins', fontsize=14, fontweight = 'normal', color='grey')

centre_circle = plt.Circle((0, 0), 0.60, fc='white', edgecolor='white')
fig.gca().add_artist(centre_circle)

ax.axis('equal')

plt.legend(["DIABETES", "NOT DIABETES"], loc= 'upper center', bbox_to_anchor=(0.1, 0.1), frameon=True)

plt.show()
```

```python
st=df['Diabetes_binary']==1
diabetes = df[st]
diabetes
```

```python
pd.crosstab(df.Veggies, df.Diabetes_binary).plot(
    kind="bar",
    figsize=(8, 6),
    color=["#FF7F7F", "#7FB3FF"],
    linewidth=1.2
)

plt.title(
    "Diabetes Frequency Based on Vegetable Consumption",
    fontsize=16,
    fontweight="bold",
    fontname="Poppins",
    color="#222831"
)
plt.xlabel(
    "Vegetable Consumption",
    fontsize=14,
    fontweight="bold",
    fontname="Poppins",
    color="#333333"
)
plt.ylabel(
    "Frequency",
    fontsize=14,
    fontweight="bold",
    fontname="Poppins",
    color="#333333"
)

plt.tight_layout()
plt.show()
```
